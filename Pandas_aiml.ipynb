```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, np.nan, 30, 22, np.nan],  # Missing values in Age
    'City': ['New York', 'Los Angeles', np.nan, 'Chicago', 'Houston'],  # Missing value in City
    'Salary': [50000, 60000, 55000, np.nan, 65000]  # Missing value in Salary
}

df = pd.DataFrame(data)
print(df)
```

```python
filterd_df=df[(df['Age']>22) &(df['Salary']> 50000)]
print(filterd_df)
```

```python
filterd_df=df[df['City'].isin(['New York','Houston'])]
print(filterd_df)
```

```python
#filter row where values are missing
missing_age= df[df['Age'].isnull()]
print(missing_age)
```

```python
non_missing_sal=df[df['Salary'].notnull()]
print(non_missing_sal)
```

```python
filterd_df= df.query('Age>22 and Salary > 50000')
print(filterd_df)
```
